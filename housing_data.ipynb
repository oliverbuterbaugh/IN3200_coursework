```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
# load data
housing = fetch_california_housing()
data = pd.DataFrame(housing.data, columns=housing.feature_names)
data['MediHouseVal'] = housing.target
```

```python
# summary statistics
print(data[['MedInc', 'MediHouseVal']].describe())
```

```python
# scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['MedInc'], data['MediHouseVal'], alpha=0.3, s=10)
plt.xlabel('Median Income ($10,000s)')
plt.ylabel('Median House Value ($100,000s)')
plt.title('Median Income vs. Median House Value')
plt.savefig('scatter_income_vs_house_value.png')
plt.show()
```

```python
# histogram
plt.figure(figsize=(8, 6))
plt.hist(data['MediHouseVal'], bins=50, edgecolor='black')
plt.xlabel('Median House Value ($100,000s)')
plt.ylabel('Frequency')
plt.title('Distribution of Median House Values')
plt.savefig('histogram_house_value.png')
plt.show()
```
